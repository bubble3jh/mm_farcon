```python
from collections import Counter
import json
import pandas as pd

def singularize(word):
    if word.endswith('s'):
        return word[:-1]
    return word

# 파일 경로
att='sexuality/gay'
file_path = f"/data1/bubble3jh/farcon/Fair-Diffusion/ieat/data/experiments/{att}/captions.json"

# 파일에서 데이터 로드
with open(file_path, 'r') as file:
    captions_data = json.load(file)

# 모든 캡션을 하나의 텍스트로 결합
all_captions = " ".join(item['caption'] for item in captions_data)

# 단어별로 분할
words = all_captions.split()

# 단어 정규화 (소문자로 변환하고 단수 형태로 변환)
normalized_words = [singularize(word.lower()) for word in words]

# 정규화된 단어의 빈도 계산
normalized_word_counts = Counter(normalized_words)

# 가장 많이 나온 단어 순위
most_common_normalized_words = normalized_word_counts.most_common(15)

# DataFrame으로 변환
df = pd.DataFrame(most_common_normalized_words, columns=['Word', 'Count'])

# 결과 표시
display(df)
```

```python
import json
from deepdiff import DeepDiff

# JSON 파일을 읽고 파이썬 객체로 변환하는 함수
def read_json(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

# 두 JSON 객체를 비교하는 함수
def compare_json_objects(obj1, obj2):
    return DeepDiff(obj1, obj2, ignore_order=True).pretty()

# 파일 경로 (여기서는 예시를 위한 가상의 경로입니다)
file_path1 = '/data1/bubble3jh/git_FarconVAE/data/mm_experiments_added/coco_captions_modified.json'
file_path2 = '/data1/bubble3jh/git_FarconVAE/data/mm_experiments_added/conception_captions_modified.json'

# 두 JSON 파일 읽기
json_obj1 = read_json(file_path1)
json_obj2 = read_json(file_path2)

# 비교 결과 출력
diff = compare_json_objects(json_obj1, json_obj2)
print(diff)
```
