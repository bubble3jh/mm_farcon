```python
from collections import Counter
import json
import pandas as pd

def singularize(word):
    if word.endswith('s'):
        return word[:-1]
    return word

# 파일 경로
att='sexuality/gay'
file_path = f"/data1/bubble3jh/farcon/Fair-Diffusion/ieat/data/experiments/{att}/captions.json"

# 파일에서 데이터 로드
with open(file_path, 'r') as file:
    captions_data = json.load(file)

# 모든 캡션을 하나의 텍스트로 결합
all_captions = " ".join(item['caption'] for item in captions_data)

# 단어별로 분할
words = all_captions.split()

# 단어 정규화 (소문자로 변환하고 단수 형태로 변환)
normalized_words = [singularize(word.lower()) for word in words]

# 정규화된 단어의 빈도 계산
normalized_word_counts = Counter(normalized_words)

# 가장 많이 나온 단어 순위
most_common_normalized_words = normalized_word_counts.most_common(15)

# DataFrame으로 변환
df = pd.DataFrame(most_common_normalized_words, columns=['Word', 'Count'])

# 결과 표시
display(df)
```

```python
import json
import os

# JSON 파일의 경로 설정
json_file_path = '/data1/bubble3jh/farcon/git_FarconVAE/data/mm/captions_transformer_coco.json'

# JSON 파일이 존재하는지 확인
if os.path.exists(json_file_path):
    # JSON 파일을 읽어들이기
    with open(json_file_path, 'r') as f:
        data = json.load(f)

# 디렉토리별로 caption 데이터를 저장할 딕셔너리
directory_captions = {}

# 데이터를 순회하면서 딕셔너리에 저장
for item in data:
    image_path = item["image_path"]
    caption = item["caption"]
    
    # 디렉토리 경로 추출
    directory_path = os.path.dirname(image_path)
    
    # 딕셔너리에 데이터 저장
    if directory_path not in directory_captions:
        directory_captions[directory_path] = []
    
    directory_captions[directory_path].append({
        "image_path": image_path,
        "caption": caption
    })

# 딕셔너리의 각 항목을 새로운 JSON 파일로 저장
for dir_path, captions in directory_captions.items():
    # 새로운 JSON 파일 경로
    new_json_path = os.path.join(dir_path, "captions.json")
    
    # JSON 파일 쓰기
    with open(new_json_path, 'w') as f:
        json.dump(captions, f, indent=4)
```
